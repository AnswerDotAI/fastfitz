# fastfitz

> Read, search, and visually preview PDFs with `fitz` (PyMuPDF), fastai-style

fastfitz adds a small ergonomic layer over [PyMuPDF](https://pymupdf.readthedocs.io/): page-tagged text extraction, document-level search returning `(page, rects)` pairs, and search-driven visual previews (`pg.preview('some clause')` renders just the matched region). Pages get a `_repr_png_`, so a bare `Page` expression displays itself in any rich frontend -- Jupyter, solveit, or an AI-driven kernel.

## Install

```sh
pip install fastfitz
```

fastfitz also registers a [pyskill](https://answerdotai.github.io/pyskills), so AI kernels with pyskills discovery gain PDF viewing automatically when fastfitz is installed.

## Quick start

````
#| eval: false
import fitz
from fastfitz.core import *

pdf = fitz.open('some.pdf')
print(pdf.text(pages=[0]))                # cheap first: page-tagged text
hits = pdf.search_for('Delaware')         # [(page, rects)] across the document
hits[0][0].preview('Delaware')            # render just the matched region
pdf[0]                                    # bare page displays itself
````
